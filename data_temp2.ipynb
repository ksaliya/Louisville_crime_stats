```python
import pandas as pd
```

```python
df = pd.DataFrame()
for year in range(2007, 2018):
    temp = pd.read_csv('crime-data-All-2018-07-22_0030/Crime_Data_{}.csv'.format(year))
    df = df.append(temp)
```

```python
df['YEAR_REPORTED'] = [df['DATE_REPORTED'].iloc[i].split('-')[0] for i in range(0, len(df))]
```

```python
df.head(2)
```

```python
df['MONTH_REPORTED'] = [df['DATE_REPORTED'].iloc[i].split('-')[1] for i in range(0, len(df))]
df['DAY_REPORTED'] = [df['DATE_REPORTED'].iloc[i].split('-')[2].split()[0] for i in range(0, len(df))]

# df['YEAR_OCCURED'] = [df['DATE_OCCURED'].iloc[i].split('-')[0] for i in range(0, len(df))]
# df['MONTH_OCCURED'] = [df['DATE_OCCURED'].iloc[i].split('-')[1] for i in range(0, len(df))]
# df['DAY_OCCURED'] = [df['DATE_OCCURED'].iloc[i].split('-')[2].split()[0] for i in range(0, len(df))]

# df['DATE_REPORTED'].iloc[0].split('-')[2].split()[0]
```

```python
df.head(2)
```

```python
name = ['YEAR_OCCURED', 'MONTH_OCCURED', 'DAY_OCCURED']
for j in range(0, len(name)):
    temp = []
    for i in range(0, len(df)):
        try:
            if j != 2:
                temp.append(df['DATE_OCCURED'].iloc[i].split('-')[j])
            else:
                temp.append(df['DATE_OCCURED'].iloc[i].split('-')[j].split()[0])
        except:
            temp.append('NaN')
    df[name[j]] = temp
```

```python
crimes_reported_by_year = pd.crosstab(df.YEAR_REPORTED,df.CRIME_TYPE)
```

```python
plt.figure(figsize=(15,10))  
for i in range(0, len(crimes_reported_by_year.columns)):
    plt.plot(crimes_reported_by_year.index.values, crimes_reported_by_year[crimes_reported_by_year.columns[i]])

plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.title('Number of crimes by category over 10 years')
plt.xlabel('Year')
plt.ylabel('Crime count')
plt.show()
```
